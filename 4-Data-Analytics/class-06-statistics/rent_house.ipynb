## OLS (Ordinary Least Squares)

```python
x = df_rents["area_m2"]
y = df_rents["valor_aluguel"]

x = sm.add_constant(x)
model = sm.OLS(y, x)
result = model.fit()
print(result.summary())
```

```python
plt.figure(figsize=(12, 8))
plt.xlabel("area_m2", size=16)
plt.ylabel("valor_aluguel", size=16)
plt.plot(x["area_m2"], y, "o", label="Dados Reais")
plt.plot(x["area_m2"], result.fittedvalues, "r-", label="Linha de regressÃ£o")
plt.legend(loc="best")
plt.show()
```
