```python
import matplotlib.pyplot as plt
# pip install matplotlib
# https://matplotlib.org
```

```python
x = [1, 2, 3]
y = [2, 5, 6]

plt.plot(x, y, label="Label")
plt.legend()
plt.xlabel("Tempo")
plt.ylabel("Valor")
plt.title("Valor x Tempo")
plt.show()
```

```python
# what is your favorite fruit?
data = [80, 20, 35, 120]
names = ["banana", "apple", "grape", "mango"]
colors = ["yellow", "red", "violet", "orange"]
plt.pie(data, labels=names, colors=colors, startangle=90, explode=(0, 0.2, 0, 0))
plt.show()
```

```python
list1 = [12, 14, 15, 27, 23, 25, 18, 22]
list2 = [22, 16, 12, 17, 26, 15, 28, 23]
list3 = [28, 13, 17, 15, 16, 25, 17, 11]
list4 = [11, 17, 21, 22, 16, 15, 18, 13]

plt.bar(list1, list2, label="Cidade1", color="orange")
plt.bar(list3, list4, label="Cidade2", color="blue")
plt.xlabel("Idade")
plt.ylabel("Milhes")
plt.legend()
plt.show()
```

```python
plt.scatter(list1, list4, label="Pontos", color="black", marker="o")
plt.xlabel("Idade")
plt.ylabel("Milhes")
plt.legend()
plt.show()
```

```python
import numpy as np
numbers = np.random.randn(100000)
plt.hist(numbers, bins=100)
plt.title('Histograma de Dados')
plt.xlabel('Valores')
plt.ylabel('Frequncia')
plt.show()
```
