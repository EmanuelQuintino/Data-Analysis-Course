```python
import pandas as pd
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
```

```python
housing = datasets.fetch_california_housing()
print(housing)
```

```python
df = pd.DataFrame(housing.data, columns=housing.feature_names)
df
```

```python
df.describe()
```

```python
df_target = pd.DataFrame(housing.target, columns=housing.target_names)
df_target
```

```python
x = df
y = df_target

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=123)
```

```python
x_train
```

```python
y_train
```

```python
scaler = StandardScaler()
x_train_scaler = scaler.fit_transform(x_train)
x_test_scaler = scaler.fit_transform(x_test)
```

```python
x_train_scaler
```

```python
reg = LinearRegression().fit(x_train_scaler, y_train)
reg
```

```python
result = reg.predict(x_test_scaler)
result
```

```python
score_r2 = reg.score(x_test_scaler, y_test)
print(f"R² Score do modelo: {score_r2}")
```

```python
new_housing = [
  [8.3252, 41.0, 6.9841, 1.02381, 322.0, 2.5556, 37.88, -122.23],
  [7.2574, 38.0, 5.1234, 1.23456, 300.0, 2.3456, 35.65, -120.15]
]

new_housing_scaler = scaler.transform(new_housing)
new_housing_predict = reg.predict(new_housing_scaler)

print(f"Previsão de preço para a Casa 1: {new_housing_predict[0]}")
print(f"Previsão de preço para a Casa 2: {new_housing_predict[1]}")
```
