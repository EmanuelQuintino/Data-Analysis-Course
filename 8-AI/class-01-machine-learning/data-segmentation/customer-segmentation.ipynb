# Class Data Segmentation

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
df_customers = pd.read_csv("datasets/dados_clientes.csv")
df_customers
```

## EDA

```python
df_customers.info()
```

```python
df_customers[["idade", "renda_anual", "pontuacao_gastos"]].describe()
```

## Data Preprocessing

```python
standard = StandardScaler()
data_standard = standard.fit_transform(df_customers[["idade", "renda_anual", "pontuacao_gastos"]])
data_standard
```

## Create Model

```python
kmeans = KMeans(n_clusters = 3)
kmeans.fit(data_standard)
df_customers["cluster"] = kmeans.labels_
df_customers
```

```python
mean_cluster = df_customers.groupby("cluster").agg({
  "id": "count", 
  "idade": "mean", 
  "renda_anual": "mean", 
  "pontuacao_gastos": "mean"
  })

mean_cluster = mean_cluster.rename(columns={"id": "count"})
mean_cluster
```
